## 1.6 The CHSH Game

* [Q# exercise: CHSH Game](./1-Basic_Quantum_Concepts/6-CHSH_Game.ipynb#qchsh)

The CHSH Game sheds more light on the nature of quantum entanglement and helps us understand that
entanglement really is "spooky action at a distance," as Albert Einstein called it. The CHSH Game is named
for four physicists who discovered the [CHSH Inequality](https://en.wikipedia.org/wiki/CHSH_inequality): John Clauser, Michael Horne, Abner Shimony, and
Richard Holt. In essence, the CHSH Game/Inequality illustrates that quantum entanglement cannot be
explained by a [local hidden-variable theory](https://en.wikipedia.org/wiki/Local_hidden-variable_theory). For quantum computing, this is another algorithm where we
can employ a feature of quantum mechanics to yield a solution that is more efficient than any classical
solution.

The game is played as follows:

![Figure CHSHGame](media/CHSH_game.png)

Alice and Bob are physically separated and not allowed to communicate with one another, during
the game. (They are allowed to set a strategy together before the game begins and share entangled qubits,
however). The Referee sends a uniformly random bit x to Alice and a uniformly random bit y to Bob.
Then, Alice and Bob send back their respective answer bits, a and b, based on the strategy they set ahead
of time.

Alice and Bob win the game if a $\oplus$ b = x $\cdot$ y.  So, if either x or y is 0, then a and b should be the
same, but if x and y are both one, then a and b should be different.

If we look at the classical solution first, suppose Alice and Bob decide on a trivial strategy of always
returning 0 for both a and b. Then, we can see that they will win at most 75% of the rounds of the game
(on average):

But if we dig into the classical case further, we can see that no matter what function of x or y that
we pick, to yield a and b, we can't win more than 75% of the time, because of the difference between the
first three rows and the last row. What if we employ a quantum strategy, based on a shared pair of
entangled qubits that are in a Bell state $\frac{|00\rangle + |11\rangle}{\sqrt 2}$?

If x = 0, then Alice measures her entangled qubit in the |0⟩ basis and if her measurement is |0⟩
then she returns a = 0 and if her measurement is |1⟩ then she returns a = 1. If x = 1, then Alice
measures her entangled qubit in the |+⟩ basis and if her measurement is |+⟩ then she returns a = 0
and if her measurement is |−⟩ then she returns a = 1. If y = 0, then Bob measures his entangled
qubit in the |x⟩ basis, rotated by $\frac{\pi}{8}$ , and if his measurement is |$\frac{\pi}{8}\rangle$ then he returns b = 0 and if his
measurement is |$\frac{-\pi}{8}\rangle$ then he returns b = 1. If y = 1, then Bob measures his entangled qubit in the
|x⟩ basis, rotated by $\frac{-\pi}{8}$; and if his measurement is |$\frac{-\pi}{8}\rangle$ then he returns b = 0 and if his measurement
is |$\frac{\pi}{8}\rangle$ then he returns b = 1, where

|$\frac{\pi}{8}\rangle = \cos (\frac{\pi}{8}) |0\rangle + \sin (\frac{\pi}{8}) |1\rangle $

|$\frac{-\pi}{8}\rangle = \cos (\frac{-\pi}{8}) |0\rangle + \sin (\frac{-\pi}{8}) |1\rangle $

Let's take a look at how well this quantum strategy works. We can look at two different cases:
Case(1) is the first three rows in the table above (where the classical a = b = 0 strategy wins all of the
time) and Case(2) is the last row in the table, where the classical strategy loses all of the time.

Case(1):

Suppose that Alice has x = 0 and she measures |0⟩ meaning that she will return a = 0. So, they will
win if Bob outputs b = 0. We know that Bob's qubit is in the |0⟩ state, because it is entangled with
Alice's which already measured |0⟩. So Bob will measure |0⟩ in a basis rotated by $\frac{\pi}{8}$, and he outputs b =
0 if he measures |$\frac{\pi}{8}\rangle$ and the probability that Bob does this is $\cos^2 (\frac{\pi}{8})$ ≈ 85%.

Case(2):

Now, Alice gets x = 1 and measures her entangled qubit in the |+⟩ basis and suppose she gets |+⟩
meaning that she will return a = 0. So, they will win if Bob outputs b = 1. We know that Bob's qubit
is in the |+⟩ state, so Bob will measure |+⟩ in a basis rotated by $\frac{-\pi}{8}$, because y = 1 in this case. So, the
probability of getting $\frac{\pi}{8}$ as a result of Bob's measurement is the angle between |+⟩ and $\frac{\pi}{8}$ which is
$\cos^2 (\frac{\pi}{8})$ ≈ 85%.

If Alice and Bob employ the quantum strategy, their chances of winning the CHSH Game improve
by 10%, going up from 75% to 85%. Those are good odds!

![Figure CHSHGameQuirk](media/CHSH_game_quirk.png)


The top wire is Alice's entangled qubit and the bottom wire is Bob's—which compute a and b,
respectively. The second and the fourth wire are the Referee's qubits, x and y, respectively. The
combination of the H gate and a measurement is a coin flip, so that is how the Referee sends random bits
to Alice and Bob. You can see that Alice's and Bob's computations don't communicate, because Alice's
and Bob's computations are carried out between the ellipses (...) on Alice's and Bob's wires and there are
no control wires linking those two wires. Finally, the controlled-Not gates after the final ellipses (...) are
how the results are scored. We see the 85% chance of winning with the Quantum solution (the classical
solution isn't represented here).
