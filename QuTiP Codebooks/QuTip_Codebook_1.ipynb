# Introduction to QuTiP

QuTiP (Quantum Toolbox in Python) is an open-source software library for simulating the dynamics of open and closed quantum systems. It provides numerical simulations of a wide range of physical systems, including quantum mechanics, quantum field theory, and quantum optics. QuTiP features a user-friendly interface and allows for the creation and manipulation of quantum states and operators, as well as the calculation of various physical observables.

### Creating and Inspecting Quantum Objects


We can create a new quantum object using the `Qobj` class constructor, like:

```python
q = Qobj([[1],[0]])
print(q)
```

What we did here was passing a python list as an argument to the `Qobj` class constructor .

This is then used to construct the matrix representation of the quantum objects, and other properties.

-----

We can inspect other properties as well:

```python
# the dimension or composite Hilbert State space structure
q.dims
```

```python
# the shape of the matrix data representation
q.shape
```

```python
# the matrix data itself in sparse matrix format
print(q.data)
```

```python
# the full dense matrix
q.full()
```

```python
# some additional properties
q.isherm, q.type
```

## Using `Qobj` instances for calculations

With `Qobj` instances we can do arithmetic and apply a number of different operations using class methods:

```python
sy = Qobj([[0,-1j], [1j,0]]) # the Pauli Y matrix operator
print(sy)
```

```python
# the Pauli Z operator

sz = Qobj([[1,0],[0,-1]])
print(sz)
```

Some arithmetic with Quantum Objects

```python
H = 1.0 * sz + 0.1 * sy
print("Qubit Hamiltonian = \n")
print(H)
```

Example of modifying quantum objects using the `Qobj` methods:

```python
# the hermitian conjugate
print(sy.dag())
```

The Trace

```python
print(H.tr())
```

The Eigen Energies

```python
H.eigenenergies()
```

## States and Operators


**States** refer to the mathematical representation of a quantum system, and are typically described by vectors in a complex Hilbert space. In QuTiP, quantum states can be represented by objects called "quantum objects."

------

**Operators**, on the other hand, describe the interactions and transformations that can occur in a quantum system. In quantum computing, operators are represented by matrices, and are used to manipulate the state of a quantum system. In QuTiP, operators can be represented by objects called "quantum operator" or "Qobj."

-------

With these two concepts, QuTiP allows for the simulation of the dynamics of quantum systems, including the calculation of observables such as probability distributions and expectation values, as well as the simulation of quantum algorithms and quantum information processing tasks.

We can use functions in QuTiP that generates common states and operators for us. Here are some examples: 

### State Vectors

In QuTiP, state vectors are represented by objects called "quantum objects" or "Qobj." These objects are used to represent the state of a quantum system, and are described by complex-valued vectors in a Hilbert space.

-------

In QuTiP, state vectors can be created from a variety of inputs, including numerical arrays, analytical functions, and density matrices. Once created, state vectors can be manipulated using various QuTiP functions and operators, such as unitary transformations, measurements, and state evolution under Hamiltonian dynamics.

-----

Additionally, QuTiP allows for the calculation of various physical observables, such as the expectation values of operators, probabilities of measurement outcomes, and entropy. These observables can provide valuable insights into the behavior of quantum systems and are essential for many quantum information processing tasks.

### Fock State

A Fock state, also known as a number state, is a specific type of quantum state in quantum optics and quantum field theory. It is a basis state that represents the state of a quantum system with a definite number of particles.

For example, in a quantum harmonic oscillator, a Fock state is described by a wavefunction that corresponds to a definite number of quanta of energy in the system. These states are usually labeled with a non-negative integer n, which corresponds to the number of quanta in the state. The Fock state |n⟩ represents the state in which n quanta are present in the system.

### Coherent state

A coherent state is a specific type of quantum state that is widely used in the study of quantum optics. It is a quantum state that resembles a classical wave and is defined as the eigenstate of the lowering operator of a quantum harmonic oscillator, such as the position or the momentum operator.

-----

Coherent states have several important properties that make them useful in the study of quantum systems. They are characterized by having a definite phase and a definite amplitude, similar to classical waves. They also minimize the uncertainty principle for the position and momentum operators, and remain close to their initial form under the time evolution governed by the Hamiltonian of the system.

---

In QuTiP, coherent states can be generated using functions that create quantum objects in the desired state. These functions allow for the specification of the amplitude and phase of the coherent state, as well as the number of quanta in the state. The properties of the coherent state, such as its mean values and uncertainties, can be calculated using various functions and methods provided by QuTiP


```python
print(coherent(N=10, alpha=1.0))
```

### Density Matrices

A density matrix is a mathematical representation of a quantum state that describes the state of a quantum system in terms of probabilities. It is a positive semi-definite, Hermitian matrix that is used to describe mixed states, which are states that cannot be described by a single pure state.

-----

In contrast to pure states, which are represented by state vectors, mixed states can be represented by probability distributions over a set of pure states. The density matrix provides a convenient way of representing these probability distributions, as it allows for the calculation of various physical observables, such as the expectation values of operators, and the entropy of the state.


-----


In QuTiP, density matrices can be represented by objects called "quantum objects" or "Qobj," and can be created from a variety of inputs, including state vectors and classical probability distributions. QuTiP provides functions for the manipulation and analysis of density matrices, such as the calculation of trace, determinant, and eigenvalues, as well as the calculation of observables such as entropy and purity.

## Composite Systems

We are mostly interested in coupled quantum systems, for example coupled qubits , a qubit coupled to a cavity(oscillator mode)


------

To define states and operators for such systems in QuTiP, we use the `tensor` function to create `Qobj` instances for the composite system

-----

Consider a system composed of two qubits. If we want to create a Pauli $\sigma_z$ operator that acts on the first qubit and leaves the second qubit unaffected (i.e the operator $\sigma_z \otimes 1$) we would do

```python
sz1 = tensor(sigmaz(), qeye(2))
print(sz1)
```

We can easily verify that this two-qubit operator does indeed have the desired properties:

```python
psi1 = tensor(basis(N,1), basis(N,0)) # exciited first qubit
psi2 = tensor(basis(N,0), basis(N,1)) # excited second qubit
```

```python
sz1 * psi1 == psi1 
```

This is false, because `sz1` should flip the sign of the excited state of `psi1`
     


```python
sz1 * psi2 == psi2
```

This should be true, because `sz1` should leave `psi2` unaffected
     


Above we used the `qeye(N)` function, which generates the identity operator with `N` quantum states. If we want to do the same thing for the second qubit we can do:

```python
sz2 = tensor(qeye(2), sigmaz())
print(sz2)
```

Note the order of the `tensor` argument to the tensor function, and the correspondingly different matrix representation of the two operators `sz1` and `sz2`.

Using the same method we can create coupling terms of the form $\sigma_x \otimes \sigma_x$
:

```python
a = tensor(sigmax(), sigmax())
print(a)
```

Now we are ready to create a `Qobj` representation of a coupled two-qubit Hamiltonian:

$$H = \epsilon_1 \sigma_z^{(1)} + \epsilon_2 \sigma_z^{(2)} + g \sigma_x^{(1)}\sigma_x^{(2)}$$


```python
epsilon = [1.0, 1.0]
g = 0.1

sz1 = tensor(sigmaz(), qeye(2))
sz2 = tensor(qeye(2), sigmaz())

H = epsilon[0] * sz1 + epsilon[1] * sz2 + g * tensor(sigmax(), sigmax())

print(H)
```

To create composite systems of different types, all we need to do is to change the operators that we pass to the `tensor` function (which can take an arbitrary number of operator for composite systems with many components).

For example, the Jaynes-Cumming Hamiltonian for a qubit-cavity system:

$$H = \omega_c a^\dagger a - \frac{1}{2}\omega_a \sigma_z + g (a \sigma_+ + a^\dagger \sigma_-)$$

```python
wc = 1.0 # cavity frequency
wa = 1.0 # qubit/atom frenqency
g = 0.1  # coupling strength

# cavity mode operator
a = tensor(destroy(5), qeye(2))

# qubit/atom operators
sz = tensor(qeye(5), sigmaz())   # sigma-z operator
sm = tensor(qeye(5), destroy(2)) # sigma-minus operator

# the Jaynes-Cumming Hamiltonian
H = wc * a.dag() * a - 0.5 * wa * sz + g * (a * sm.dag() + a.dag() * sm)

print(H)
     
```

Note that:

$$a \sigma_+ = (a \otimes \mathbf{1}) (\mathbf{1} \otimes \sigma_+)$$

So the following two are identical:

```python
a = tensor(destroy(3), qeye(2))
sp = tensor(qeye(3), create(2))

print(a * sp)
     
```

## Unitary Dynamics

Unitary evolution of a quantum system in QuTiP can be calculated with the `mesolve` function.

----

`mesolve` is short for Master-eqaution solve (for dissipative dynamics), but if no collapse operators (which describe the dissipation) are given to the solve it falls back on the unitary evolution of the Schrodinger (for initial states in state vector for) or the von Neuman equation (for initial states in density matrix form).

-----

The evolution solvers in QuTiP returns a class of type Odedata, which contains the solution to the problem posed to the evolution solver.

-----

For example, considor a qubit with Hamiltonian $H = \sigma_x$
 and initial state $|1⟩$
 (in the sigma-z basis): Its evolution can be calculated as follows:

```python
# Hamiltonian
H = sigmax()

# initial state
psi0 = basis(2, 0)

# list of times for which the solver should store the state vector
tlist = np.linspace(0, 10, 100)

result = mesolve(H, psi0, tlist, [], [])
```

```python
result
```

The `result` object contains a list of the wavefunctions at the times requested with the `tlist` array.

```python
len(result.states)
     
```

```python
print(result.states[-1] )

# the final state
```

## Expectation Values

The expectation values of an operator given a state vector or density matrix (or list thereof) can be calculated using the `expect` function.

```python
expect(sigmaz(), result.states[-1])
```

```python
expect(sigmaz(), result.states)
```

```python
fig, axes = plt.subplots(1,1)

axes.plot(tlist, expect(sigmaz(), result.states))

axes.set_xlabel(r't', fontsize=20)
axes.set_ylabel(r'sigma_z', fontsize=20);
```

If we are only interested in expectation values, we could pass a list of operators to the `mesolve` function that we want expectation values for, and have the solver compute then and store the results in the `Odedata` class instance that it returns.

For example, to request that the solver calculates the expectation values for the operators:

$\sigma_x, \sigma_y, \sigma_z$

```python
result = mesolve(H, psi0, tlist, [], [sigmax(), sigmay(), sigmaz()])
```

Now the expectation values are available in `result.expect[0]`, `result.expect[1]`, and `result.expect[2]`:

```python
fig, axes = plt.subplots(1,1)

axes.plot(tlist, result.expect[2], label=r'sigma_z')
axes.plot(tlist, result.expect[1], label=r'sigma_y')
axes.plot(tlist, result.expect[0], label=r'sigma_x')

axes.set_xlabel(r't', fontsize=20)
axes.legend(loc=2);
```

## Dissipative Dynamics

To add dissipation to a problem, all we need to do is to define a list of collapse operators to the call to the `mesolve` solver.

A collapse operator is an operator that describes how the system is interacting with its environment.

----


For example, consider a quantum harmonic oscillator with Hamiltonian

$$ H = \hbar\omega a^\dagger a $$

and which loses photons to its environment with a relaxation rate 
$\kappa$. The collapse operator that describes this process is

$$ \sqrt{\kappa} a$$

since $a$
 is the photon annihilation operator of the oscillator.

To program this problem in QuTiP:




```python
w = 1.0               # oscillator frequency
kappa = 0.1           # relaxation rate
a = destroy(10)       # oscillator annihilation operator
rho0 = fock_dm(10, 5) # initial state, fock state with 5 photons
H = w * a.dag() * a   # Hamiltonian

# A list of collapse operators
c_ops = [np.sqrt(kappa) * a]
```

```python
tlist = np.linspace(0, 50, 100)

# request that the solver return the expectation value of the photon number state operator a.dag() * a
result = mesolve(H, rho0, tlist, c_ops, [a.dag() * a]) 
     
```

```python
fig, axes = plt.subplots(1,1)
axes.plot(tlist, result.expect[0])
axes.set_xlabel(r't', fontsize=20)
axes.set_ylabel(r"Photon number", fontsize=16);
```
